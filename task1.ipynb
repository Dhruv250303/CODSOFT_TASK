```python
import datetime
import time

class Task:
    def _init_(self, title, description, due_date, reminder_time):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.reminder_time = reminder_time
        self.completed = False

tasks = []

def add_task():
    title = input("Enter task title: ")
    description = input("Enter task description: ")
    due_date_str = input("Enter due date (YYYY-MM-DD HH:MM): ")
    reminder_time_str = input("Enter reminder time (YYYY-MM-DD HH:MM): ")

    due_date = datetime.datetime.strptime(due_date_str, "%Y-%m-%d %H:%M")
    reminder_time = datetime.datetime.strptime(reminder_time_str, "%Y-%m-%d %H:%M")

    task = Task(title, description, due_date, reminder_time)
    tasks.append(task)
    print(f"Task '{title}' added.")

def update_task():
    display_tasks()
    task_index = int(input("Enter the task number to update: ")) - 1

    if 0 <= task_index < len(tasks):
        task = tasks[task_index]
        task.title = input(f"Enter new title (current: {task.title}): ") or task.title
        task.description = input(f"Enter new description (current: {task.description}): ") or task.description
        due_date_str = input(f"Enter new due date (current: {task.due_date}) (YYYY-MM-DD HH:MM): ")
        reminder_time_str = input(f"Enter new reminder time (current: {task.reminder_time}) (YYYY-MM-DD HH:MM): ")

        if due_date_str:
            task.due_date = datetime.datetime.strptime(due_date_str, "%Y-%m-%d %H:%M")
        if reminder_time_str:
            task.reminder_time = datetime.datetime.strptime(reminder_time_str, "%Y-%m-%d %H:%M")

        print(f"Task '{task.title}' updated.")
    else:
        print("Invalid task number.")

def delete_task():
    display_tasks()
    task_index = int(input("Enter the task number to delete: ")) - 1

    if 0 <= task_index < len(tasks):
        task = tasks.pop(task_index)
        print(f"Task '{task.title}' deleted.")
    else:
        print("Invalid task number.")

def mark_task_completed():
    display_tasks()
    task_index = int(input("Enter the task number to mark as completed: ")) - 1

    if 0 <= task_index < len(tasks):
        task = tasks[task_index]
        task.completed = True
        print(f"Task '{task.title}' marked as completed.")
    else:
        print("Invalid task number.")

def display_tasks():
    if not tasks:
        print("No tasks available.")
        return

    for idx, task in enumerate(tasks, 1):
        status = 'Completed' if task.completed else 'Pending'
        print(f"{idx}. {task.title} - {task.description} (Due: {task.due_date}, Reminder: {task.reminder_time}) - {status}")

def check_reminders():
    current_time = datetime.datetime.now()
    for task in tasks:
        if not task.completed and current_time >= task.reminder_time:
            print(f"Reminder: {task.title} - {task.description} is due at {task.due_date}!")

def main():
    while True:
        print("\nTo-Do List Application")
        print("1. Add Task")
        print("2. Update Task")
        print("3. Delete Task")
        print("4. Mark Task Completed")
        print("5. Display Tasks")
        print("6. Exit")
        
        choice = input("Choose an option: ")

        if choice == '1':
            add_task()
        elif choice == '2':
            update_task()
        elif choice == '3':
            delete_task()
        elif choice == '4':
            mark_task_completed()
        elif choice == '5':
            display_tasks()
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")

        check_reminders()
        time.sleep(1)

if "_main_" == "_main_":
    main()
```
